```python
#Promedio de edad de las mujeres que sobrevivieron.
dfI = df.loc[df['sex'] == 'female']
dfI.loc[dfI['survived'] == 1,['age']].mean()
```

```python
#Cantidad de los hombres de 3er clase que sobrevivieron.
dfII = df.loc[df['sex'] == 'male']
dfII.loc[(dfII['class'] == 'Third') & (dfII['survived'] == 1),['age']].count()
```

```python
#El sobreviviente hombre de mayor edad que sobrevivio y que pertencia a tercera clase
dfIII = df.loc[df['survived'] == 1]
dfIII.loc[(dfIII['class'] == 'Third') & (dfIII['sex'] == 'male'),['age']].max()
```
