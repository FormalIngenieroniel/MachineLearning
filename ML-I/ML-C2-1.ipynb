Se identifican variables numericas
 A la columna Income, se le elimina el signo "$" ademas de el "." para convertir a float sin ambiguedad

```python
# prompt:  A la columna Income, se le elimina el signo "$" ademas de el "." para convertir a float sin ambiguedad
# Reemplazar '$' y '.' por '' en la columna Income
df['Income'] = df['Income'].str.replace('$', '', regex=False).str.replace('.', '', regex=False)

# Convertir la columna Income a tipo float
df['Income'] = pd.to_numeric(df['Income'], errors='coerce')

# Mostrar las primeras filas para verificar
df.head()
```

```python
# prompt: A la columna limit, se le elimina el signo "$" ademas de convertir a float sin ambiguedad

import pandas as pd
df['Limit'] = df['Limit'].str.replace('$', '', regex=False)
df['Limit'] = pd.to_numeric(df['Limit'], errors='coerce')

# Mostrar las primeras filas para verificar
df.head()
```

Se modifican las variables categoricas (own, student y ,married) con 1 o 0 para evitar variables adicionales innecesarias, se modifica region con nominal

```python
df['Own'] = df['Own'].map({'Yes': 1, 'No': 0})
df['Student'] = df['Student'].map({'Yes': 1, 'No': 0})
df['Married'] = df['Married'].map({'Yes': 1, 'No': 0})

df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse_output=False)
region_encoded = encoder.fit_transform(df[['Region']])
region_encoded_df = pd.DataFrame(region_encoded, columns=encoder.get_feature_names_out(['Region']))

df = pd.concat([df, region_encoded_df], axis=1)
df = df.drop('Region', axis=1)

df.head()
```

Se crea un pipeline numerico con un simple imputer y un standar scaler

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler

numeric_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])
```

Se crea un pipeline nominal con un simple imputer y un ordinal encoder

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.impute import KNNImputer

nominal_pipeline = Pipeline([
    ('imputer', KNNImputer(n_neighbors=3)),
    ('encoder', OneHotEncoder())
])

ordinal_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('encoder', OrdinalEncoder())
])
```

Crear un Column transform que contenga todos los pipelines


```python
from sklearn.compose import ColumnTransformer

t = ColumnTransformer([
    ('numeric', numeric_pipeline, ['Income', 'Limit', 'Rating', 'Cards', 'Age', 'Education']),
    ('nominal', nominal_pipeline, ['Student', 'Married', 'Own','Region_East']),
    ('ordinal', ordinal_pipeline, [])
])

t
```

```python
df_transformed = t.fit_transform(df)
pd.DataFrame(df_transformed)
```

Se va acrear otro pipeline donde el primer paso sera el proprocessor (la union de los anteriores piplenes) seguido del estimador y un knnRegresor()

```python
from sklearn.neighbors import KNeighborsRegressor

pipeline_modelo = Pipeline([
    ('preprocessor', t),
    ('modelo', KNeighborsRegressor())
])

pipeline_modelo
```
